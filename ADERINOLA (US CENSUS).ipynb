### Question 1: State with the most countries in it

```python
M= C.groupby('STNAME')['COUNTY'].nunique().sort_values(ascending=False).head(5)
M
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Create a horizontal bar chart
plt.bar(M.index, M.values)

# Add data labels to the chart
for i, v in enumerate(M.values):
    plt.text(i, v + 0.25, str(v), color='red', fontweight='bold', ha='center')

# Set the title and axis labels
plt.title('State with most countries')
plt.xlabel('STATE')
plt.ylabel('COUNTRY')

# Rotate the x-axis labels
plt.xticks(rotation=45)

# Display the chart
plt.show()
```

##### Texas has the high Countries of 255. So, TEXAS is the state with the most countries
